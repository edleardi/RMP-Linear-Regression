# Selenium scrapes ratemyproffessors for urls

```python
# Make sure to not rerun until export to csv is complete

prof_urls = {}
running_prof_count = 0
```

```python
# Run on first generation of browser

browser=webdriver.Firefox(executable_path='/Applications/geckodriver')
browser.get('http://www.ratemyprofessors.com/campusRatings.jsp?sid=1')
browser.refresh()
try:
    cookies_popup = browser.find_element_by_class_name("balaCloseImage")
    cookies_popup.click()
except:
    pass
```

```python
# Reset school_id if scraping is interuptted to the len of prof_urls

generic_url = 'http://www.ratemyprofessors.com/campusRatings.jsp?sid='
school_id = 0

# Max school_id is 6049
for i in range(6049):
    school_id += 1
    school_url = generic_url + str(school_id)
    browser.get(school_url) # Navigate to the college overview page
    
# Click on the "View All Professors" button
    try:
        top_profs = browser.find_element_by_class_name("button-link")
        top_profs.click()
        htmlElem = browser.find_element_by_class_name('left-panel')
    except:
        print("\nFailed to click View All Professors, 404 Error?")
        print("School id that failed: " + str(school_id))
# Scroll down on left panel, click load more button
    Result = None
    while Result is None:
        try:
            htmlElem.send_keys(Keys.END)
            time.sleep(3)
            # Find the load more button and click it
            load_more = browser.find_elements_by_class_name("content")[1]
            load_more.click()
        except:
            Result = "\nProfs loaded from school id: " + str(school_id)
            print(Result)
            
# Extract list of every href in each li tag with id='my-professor-...'
    Result2 = None
    prof = -1
    professor_list = []
    while Result2 is None:
        try:
            prof += 1
            my_prof_url = browser.find_elements_by_xpath("//li[starts-with(@id, 'my-professor-')]/a")[prof].get_attribute('href')
            professor_list.append(my_prof_url)
        except:
            prof_urls[school_id] = professor_list
            Result2 = "Prof urls gathered from school id: " + str(school_id)
            print(Result2)
            print("School id " + str(school_id) + " gathered: " + str(len(prof_urls[school_id])) + " profs")
            prof_count = len(prof_urls[school_id])
            running_prof_count += prof_count
            print("Running prof count: " + str(running_prof_count))

for i in range(5):
    os.system("printf '\a'") # or '\7'      

```

```python
# Export out the dictionary to a csv

with open('/data/my_dictionary.csv', 'w') as csv_file:
    writer = csv.writer(csv_file)
    for school in prof_urls:
        for url in prof_urls[school]:
            writer.writerow([school, url])
    
```
